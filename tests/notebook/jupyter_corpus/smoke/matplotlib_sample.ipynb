# Matplotlib sample

Exercises `%matplotlib inline` magic (dropped on import) and figure capture.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```
